## 1.Determine the top three most common cuisines in Dataset.

```python
a =df['Cuisines'].value_counts()
top_three=a.head(3)
count_cuisines=a
```

```python
print("The top three Cuisines:")
for i in range(len(top_three.index)):
    print(top_three.index[i])
```

```python
plt.figure(figsize=(7,5))
color=['pink','brown','brown']
top_three.plot(kind='bar',color=color)
plt.title('Top three most visited Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Number of Restaurants')
plt.show()
```

## Percentage of top three cusines

# 3. Creating a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
